```python
import pandas as pd
import numpy as np
df1=pd.read_excel("amz_com-ecommerce_sample.xlsx")
df1.head()
```

```python
import pandas as pd
import numpy as np
df2=pd.read_excel("flipkart_com-ecommerce_sample.xlsx")
df2.head()
```

```python
dpid1={}
dpid2={}
dpdtname1={}
dpdtname2={}
for i in df1['pid']:
    if i not in dpid1:
        dpid1[i]=1
    else:
        dpid1[i]=dpid1[i]+1
for i in df2['pid']:
    if i not in dpid2:
        dpid2[i]=1
    else:
        dpid2[i]=dpid2[i]+1
for i in df1['product_name']:
    if i not in dpdtname1:
        dpdtname1[i]=1
    else:
        dpdtname1[i]=dpdtname1[i]+1
for i in df2['product_name']:
    if i not in dpdtname2:
        dpdtname2[i]=1
    else:
        dpdtname2[i]=dpdtname2[i]+1
print('Amazon products:',dpdtname1)
print('Flipkart products:',dpdtname2)
print('Amazon product ids:',dpid1)
print('Flipkart products ids:',dpid2)
```

```python
# From observation the important features we can use for solving the problem are "pid" and "product name" we will use "pid" for 
# selection of similar products as it the "product name" varies in amazon and flipkart data set but the "pid" is unique of a
# particular product
l=["uniq_id","crawl_timestamp","product_url","product_name","product_category_tree","pid","retail_price","discounted_price","image,is_FK_Advantage_product","description","product_rating","overall_rating","brand","product_specifications"]
```

```python
# Modifications in datafarme df1 and df2 according to our needs
df1.rename(columns = {'product_name':'Product Name In Amazon', 'retail_price':'Retail Price In Amazon','discounted_price':'Discounted Price In Amazon'}, inplace = True)
df2.rename(columns = {'product_name':'Product Name In Flipkart', 'retail_price':'Retail Price In Flipkart','discounted_price':'Discounted Price In Flipkart'}, inplace = True)
# df1.drop(df1.columns[[0,1,2,4,8,9,10,11,12,13,14]], axis=1, inplace=True)
# df2.drop(df2.columns[[0,1,2,4,8,9,10,11,12,13,14]], axis=1, inplace=True)
#merging the Dataframes and displaying it in the required order
df=pd.merge(df1,df2,on="pid")
# df1.merge(df2[['pid']])
# display(df)

df
```
